Modelo de Machine Learning desenvolvido para a recomendação de filmes

```python
#importando as bibliotecas
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#carregando o dataframe dando nome as colunas
columnsName = ['user_id','item_id','rating','timeStamp'] 
df = pd.read_csv('u.data', sep='\t',names =columnsName )
```

```python
df.head(10)
```

```python
df.shape
```

```python
#verificando dados faltantes
df.isnull().count()
```

```python
# a coluna item_id mostra o id de cada filme, vamos acrescentar no dataframe os nomes do filmes apartir desta coluna
movieIdTitle = pd.read_csv('Movie_Id_Titles') 
```

```python
movieIdTitle.head(10)
```

```python
df= pd.merge(df, movieIdTitle ,on='item_id')
```

```python
df.head(10)
```

### <p style="color:blue"> Análise exploratório dos dados </p>
Vamos analisar os dados através de plotagem de gráficos para identificar os filmes mais bem classificados 

```python
#este dataframe mostra os filmes agrupados de acordo com a média dos rating
pd.DataFrame(df.groupby('title')['rating'].mean().sort_values(ascending=False)).head()
```

```python
#este dataframe mostra a quantidade de vezes que cada filme aparece no dataset
pd.DataFrame(df.groupby('title')['rating'].count().sort_values(ascending=False)).head()
```

```python
#vamos criar um dataframe com estas informações para podermos analisar os dados através de plotes
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
```

```python
ratings.head()
```

```python
# excluindo alguns filmes (que provavelmente foram visto somente uma vez) temos uma distribuição
# normal dos filmes
plt.figure(figsize=(12,8)) 
ratings["rating"].hist(bins=30)
```

```python
#poucos filmes foram assistidos por muitos usuários
plt.figure(figsize=(12,8)) 
ratings["num of ratings"].hist(bins=30)
```

```python
#verificando se os filmes com maior pontuação possuem muitas ocorrencias no conjunto de dados
# um aumento na contagem dos filmes também gera um aumento na avaliação dos filmes
sns.jointplot(x='rating',y='num of ratings', data= ratings, alpha=0.4, size= 8)
```

### <p style="color:blue"> Recomendando filmes semelhantes </p>

##### Vamos fazer um dataframe onde as colunas são os nomes dos filmes e as linhas são todos os usuários
##### como cada usuário não assistiu a todos os filme teremos valores preenchidos com NaN

```python
movieMat = df.pivot_table(index="user_id",columns= "title", values= "rating")
```

```python
movieMat.head()
```

```python
#Os top 5 filmes mais avaliados e escolher dois deles para fazer o modelo de recomendação
```

```python
ratings.head()
```

```python
ratings.sort_values('num of ratings', ascending=False).head(5)
```

```python
# Vamos escolher star Wars e Liar Liar(um de ficção e outro de comédia)
```

```python
# Vamos pegar as avaliações dos usuários para estes dois filmes
avalStarWars = movieMat['Star Wars (1977)']
avalLiarLiar = movieMat['Liar Liar (1997)']
```

```python
# vamos obter as correlações entre essas duas series no pandas
corAvalStarWars = movieMat.corrwith(avalStarWars)
corAvalLiarLiar = movieMat.corrwith(avalLiarLiar)
```

```python
# vamos limpar NaN os dados e transformar as series em dataframes
df_corAvalStarWars = pd.DataFrame(corAvalStarWars,columns=['Correlation'])
df_corAvalStarWars.dropna(inplace=True)
df_corAvalLiarLiar = pd.DataFrame(corAvalLiarLiar, columns=['Correlation'])
df_corAvalLiarLiar.dropna(inplace=True)
```

```python
# Agora vamos ordenar os nomes dos filmes de forma decrescente das correlações, assim poderemos
# visualizar os filmes que possuem maior similaridade uns com os outros
# correlação 1: Correlacionados positivamente
# correlação 0: Não correlacionados
# correlação -1: Correlacionados negativamente
df_corAvalStarWars.sort_values('Correlation', ascending=False).head(10)
```

```python
df_corAvalLiarLiar.sort_values('Correlation',ascending=False).head(10)
```

```python
# Podemos observar que filmes que não possuem similaridade estão muito correlacionados. Isso se deve ao
# fato de que alguns destes filmes não foram assistidos por muitos usuários. Para solucionar este proble-
# ma vamos limpar os dados filtrando os filmes que possuem menos que 100 comentários
```

```python
# vamos precisar da coluna 'num of ratings' de rantings
df_corAvalStarWars = df_corAvalStarWars.join(ratings['num of ratings'])
df_corAvalLiarLiar = df_corAvalLiarLiar.join(ratings['num of ratings'])
```

```python
# Agora vamos filtrar os filmes pela quantidade de comentários
df_corAvalStarWars[df_corAvalStarWars['num of ratings'] > 100].sort_values('Correlation',ascending = False).head()
```

```python
df_corAvalLiarLiar[df_corAvalLiarLiar['num of ratings']> 100].sort_values('Correlation', ascending=False).head()
```

### <p style="color:blue"> Agora observando a correlação entre os filmes faz mais sentido, uma vez que os filmes estão na mesma categoria. Desta forma conseguimos recomendar filmes para os usuários que assistiram Star Wars e Liar Liar. Pela correlação podemos verificar se os filmes serão uma boa sujestão (quanto maior o valor da correlação mais chances do usuário gostar do filmes recomendado) </p>
